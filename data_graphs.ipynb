## Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re, string, nltk
from nltk.tokenize import word_tokenize
```

```python
# read in preprocessed training data if necessary
train_df = pd.read_csv('preprocessed_training_data.csv', sep=';', encoding='utf-8')

train_df.loc[(train_df.label == 0),'label']='World'
train_df.loc[(train_df.label == 1),'label']='Sports'
train_df.loc[(train_df.label == 2),'label']='Business'
train_df.loc[(train_df.label == 3),'label']='Sci/Tech'

train_df['label']= train_df['label'].astype(str)
```

```python
def class_distribution(df_data):
    
    # checking class distribution
    plt.figure(figsize=(10,5))

    ## train data
    plt.subplot(1,2,1)
    train_df_target = df_data['label']
    class_dist = pd.Series(train_df_target).value_counts()
    plt.title('train_df')
    plt.bar(class_dist.index, class_dist)
    plt.tight_layout()


    plt.show()
```

```python
def plot_article_length(df_data):
    
    all_cats = df_data
    len_arcticles = [len(word_tokenize(text)) for text in all_cats['text']]
    
    fig_1, ax_1 = plt.subplots(figsize=(8, 6));
    ax_1.hist(len_arcticles, bins=40, label='All Categories');
    ax_1.set_xlabel("Length of Article Descriptions");
    ax_1.set_ylabel("Number of Articles");
    
    df_sport = df_data.loc[df_data['label'] == 'Sports']
    df_world = df_data.loc[df_data['label'] == 'World']
    df_bus = df_data.loc[df_data['label'] == 'Business']
    df_tech = df_data.loc[df_data['label'] == 'Sci/Tech']
    
    
    
    fig_2, axs = plt.subplots(2, 2, figsize=(16,12))
    
    axs[0, 0].hist([len(word_tokenize(text)) for text in df_sport['text']], bins=40, label='Sport', color="red");
    axs[0, 1].hist([len(word_tokenize(text)) for text in df_world['text']], bins=40, label='World', color="green");
    axs[1, 0].hist([len(word_tokenize(text)) for text in df_bus['text']], bins=40, label='Business', color="black");
    axs[1, 1].hist([len(word_tokenize(text)) for text in df_tech['text']], bins=40, label='Sci/Tech', color="orange");
    
    axs[0, 0].set_xlabel('Length of Article Descriptions')
    axs[0, 0].set_ylabel('Number of Articles')

    axs[0, 1].set_xlabel('Length of Article Descriptions')
    axs[0, 1].set_ylabel('Number of Articles')
    
    axs[1, 0].set_xlabel('Length of Article Descriptions')
    axs[1, 0].set_ylabel('Number of Articles')
    
    axs[1, 1].set_xlabel('Length of Article Descriptions')
    axs[1, 1].set_ylabel('Number of Articles')

    ax_1.set_xlabel("Length of Article Descriptions");
    ax_1.set_ylabel("Number of Articles");

    ax_1.legend()
    axs[0, 0].legend()
    axs[0, 1].legend()
    axs[1, 0].legend()
    axs[1, 1].legend()

    
    
    
    return
```

```python
class_distribution(train_df)
```

```python
plot_article_length(train_df)
```
